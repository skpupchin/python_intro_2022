```python
from PIL import Image
import numpy as np

def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray

img = Image.open("Cat.jpg")

x = np.array(img, dtype = np.float32)
#rint(Image.fromarray(np.asarray(x, dtype=np.uint8)))
y = rgb2gray(x)

U, S, VT = np.linalg.svd(y, full_matrices=False)
print(S)
Y_full = U.dot(np.diag(S)).dot(VT)
```

```python
Image.fromarray(np.asarray(rgb2gray(x), dtype=np.uint8))
```

```python
Image.fromarray(np.asarray(Y_full, dtype=np.uint8))
```

```python
for r in range (1, 100, 10):
    Y_r = U[:,:r].dot(np.diag(S[:r])).dot(VT[:r,:])
    Y_r[Y_r < 0] = 0
    Y_r[Y_r > 255] = 255
    Image.fromarray(np.asarray(Y_r, dtype=np.uint8)).save(f'{r}.cat.jpg')
    
```
